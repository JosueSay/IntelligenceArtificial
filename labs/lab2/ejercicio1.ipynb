```python
import numpy as np             # cálculo numérico
import scipy as sp
import scipy.stats as st       # estadística

import matplotlib.pyplot as plt  # graficación
import seaborn as sns            # graficación
```

1. Obtener, mediante repeticiones, una estimación de la densidad del número de lanzamientos necesarios para obtener el
primer éxito. Por ejemplo, simular un experimento de estos N veces. Usar N = 1000.

```python
np.random.seed(12345)
p = 0.65
N = 1000
results = []

# Iterar N veces para encontrar el primer exito y guardarlo para graficar densidad
for _ in range(N):
  dist = st.bernoulli.rvs(p, size=1000)
  first_success = np.where(dist == 1)[0][0] + 1
  results.append(first_success)
  
# Graficar la distribución de los intentos necesarios
plt.hist(results, density=True, edgecolor='black')
plt.xlabel('Intentos hasta el primer éxito')
plt.ylabel('Densidad')
plt.title(f'Distribución del número de intentos hasta el primer éxito (p={p})')
plt.show()
```

2. Elaborar 3 visualizaciones de la función de densidad o masa, y cambiando los valores de p

```python
p_values = [0.65, 0.15, 0.98]

for i in range(0, len(p_values)):
    samples = st.bernoulli.rvs(p_values[i], size=1000)
    plt.figure()
    plt.hist(samples)
    plt.show()

    print(f"Usando p = {p_values[i]} con N = 1000 se obtiene que el número total de fracasos en el experimento es de {(samples==0).sum()}")
    print(f"Usando p = {p_values[i]} con N = 1000 se obtiene que el número total de éxitos en el experimento es de {(samples==1).sum()}")
    
```
